```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
# 1. Bộ lọc làm mờ (Averaging Filter)
img = cv2.imread('../anh_mau.jpg')
# Áp dụng bộ lọc làm mờ với kích thước kernel 10x10
# sử dung hàm blur của opencv
blur_img = cv2.blur(img, (10, 10))
# 2. Bộ lọc tùy chỉnh (Custom Filter)
kernel = np.ones((5, 5), np.float32) / 25
dst = cv2.filter2D(img, -1, kernel) 
# tại sao phải chia 25? bởi vì nếu tootal giá trị trong kernel lớn hơn 1 thì ảnh sau khi lọc sẽ sáng hơn ảnh gốc

# Hiển thị kết quả
plt.subplot(131), plt.imshow(img)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(blur_img)
plt.title('Blurred Image'), plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(dst)
plt.title('Filtered Image'), plt.xticks([]), plt.yticks([])


plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

# 1. Đọc ảnh
img = cv2.imread('../anh_mau.jpg')


img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # 3. Định nghĩa Sharpening Kernel (Ma trận làm nét)
    # Tổng các giá trị trong ma trận này = 5 + (-1)*4 = 1
    # Điều này giúp giữ nguyên độ sáng trung bình của ảnh
sharpening_kernel = np.array([[0, -1, 0],
                                  [-1, 5, -1],
                                  [0, -1, 0]], dtype=np.float32)

    # 4. Áp dụng bộ lọc
    # tham số thứ 2 là ddepth: -1 nghĩa là giữ nguyên độ sâu màu (như ảnh gốc)
filtered_img = cv2.filter2D(img_rgb, -1, sharpening_kernel)

    # 5. Hiển thị kết quả
plt.figure(figsize=(10, 5)) # Thiết lập kích thước khung hình cho dễ nhìn

plt.subplot(121)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(filtered_img)
plt.title('Sharpened Image')
plt.xticks([]), plt.yticks([])

plt.show()
```
