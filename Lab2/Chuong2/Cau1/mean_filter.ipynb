# Lọc trung bình (Mean Filter)

Làm mờ ảnh bằng cách lấy trung bình các điểm ảnh lân cận.


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg", 0)

if img is None:
    raise FileNotFoundError("Không tìm thấy image.jpg")

plt.imshow(img, cmap="gray")
plt.title("Ảnh xám gốc")
plt.axis("off")
```

```python
k = 5

mean_filtered = cv2.blur(img, (k, k))

plt.imshow(mean_filtered, cmap="gray")
plt.title(f"Ảnh sau lọc trung bình (Kernel {k}x{k})")
plt.axis("off")
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("Ảnh gốc")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(mean_filtered, cmap="gray")
plt.title("Ảnh sau lọc trung bình")
plt.axis("off")

plt.show()
```
