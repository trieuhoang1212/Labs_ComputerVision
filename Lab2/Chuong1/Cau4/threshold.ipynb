# Cắt ngưỡng ảnh (Thresholding)

Chuyển ảnh xám thành ảnh nhị phân bằng cách so sánh
giá trị mỗi điểm ảnh với một ngưỡng xác định.

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh xám
img = cv2.imread("image.jpg", 0)

if img is None:
    raise FileNotFoundError("The image.jpg was not found")

plt.imshow(img, cmap="gray")
plt.title("grayscale imnage")
plt.axis("off")
```

```python
# Cắt ngưỡng
T = 80

_, binary = cv2.threshold(img, T, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap="gray")
plt.title(f"Ảnh nhị phân (Threshold = {T})")
plt.axis("off")
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img, cmap="gray")
plt.title("grayscale imnage")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(binary, cmap="gray")
plt.title("Ảnh nhị phân")
plt.axis("off")

plt.show()
```
