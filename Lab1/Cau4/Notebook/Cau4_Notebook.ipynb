```python
# ct v thay di kch thc nh s dng OpenCV
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../input.jpg')

# thay i kch thc
(h, w, d) = img.shape

# tnh t l mi v kch thc mi
r = 300.0 / w 
h = 200.0 / r
dim = (300, int(h * r))

resized = cv2.resize(img, dim)
cv2.imwrite('resized_image_opencv.jpg', resized)

img_resized = cv2.imread('resized_image_opencv.jpg')

# Ct hnh nh
roi = img[50:1500, 200:1500] # ct t y=50 n y=1500 v x=200 n x=1500
cv2.imwrite('cropped_image_opencv.jpg', roi)

# hin th nh gc v nh  thay i kch thc
fig = plt.subplots(1, 3, figsize=(10, 5))
axx = fig[1]
axx[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axx[0].set_title('nh gc')
axx[1].imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
axx[1].set_title('nh thay i kch thc')
axx[2].imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
axx[2].set_title('nh  ct')
plt.show()


```

```python
# Ct v thay i kch thc nh s dng PIL
from PIL import Image
from matplotlib import pyplot as plt

img = Image.open('../input.jpg')

# Thay i kch thc v lu nh
resized_img = img.resize((600, 500))   


# Ct nh v lu nh
cropped_img = img.crop((1000, 300, 3000, 2000))  # (left, upper, right, lower)
cropped_img.save('cropped_image_PIL.jpg')

# hin th nh gc v nh  thay i kch thc
fig = plt.subplots(1,3, figsize=(10,5))
axx = fig[1]
axx[0].imshow(img)
axx[0].set_title('nh gc')
axx[1].imshow(resized_img)
axx[1].set_title('nh thay i kch thc')
axx[2].imshow(cropped_img)
axx[2].set_title('nh  ct')
plt.show()
```
