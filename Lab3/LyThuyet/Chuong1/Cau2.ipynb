## **CÁC THAM SỐ CỦA THUẬT TOÁN VÀ ẢNH HƯỞNG CỦA CHÚNG:**

# a. Tham số σ (Sigma) trong bộ lọc Gaussian

Trong thuật toán Canny, ảnh đầu vào được làm trơn bằng Gaussian filter để giảm nhiễu trước khi phát hiện biên.

Công thức của bộ lọc Gaussian 2D:

$$G(x, y) = \frac{1}{2\pi\sigma^2} \exp\left(-\frac{x^2 + y^2}{2\sigma^2}\right)$$

Ảnh hưởng của σ:

- σ nhỏ

    - Làm mờ ít

    - Giữ được nhiều chi tiết nhỏ

    - Nhạy với nhiễu

    - Phát hiện nhiều biên (kể cả biên giả)

- σ lớn

    - Làm mờ mạnh

    - Giảm nhiễu tốt

    - Có thể làm mất các biên mảnh hoặc chi tiết nhỏ

    - Biên phát hiện ít hơn nhưng ổn định hơn

* Kết luận:
Giá trị σ quyết định mức độ làm mịn ảnh, ảnh hưởng trực tiếp đến số lượng và chất lượng biên được phát hiện.

# b. Ngưỡng thấp (Low Threshold) và ngưỡng cao (High Threshold)

Sau khi tính gradient và áp dụng Non-Maximum Suppression, thuật toán Canny sử dụng Hysteresis Thresholding với hai ngưỡng:

- Ngưỡng cao (T<sub>high</sub>): xác định biên mạnh

- Ngưỡng thấp (T<sub>low</sub>): xác định biên yếu

Nguyên lý hoạt động:

- Pixel có gradient ≥ T<sub>high</sub> → biên mạnh

- Pixel có gradient giữa T<sub>low</sub> và T<sub>high</sub> → biên yếu

- Biên yếu chỉ được giữ lại nếu nối với biên mạnh

- Pixel < T<sub>low</sub> → loại bỏ

Ảnh hưởng của các ngưỡng:

T<sub>high</sub> cao

- Chỉ giữ biên rất rõ

- Có thể bỏ sót biên quan trọng

T<sub>high</sub> thấp

- Nhiều biên được phát hiện

- Dễ xuất hiện nhiễu

T<sub>low</sub> quá thấp

- Giữ nhiều biên yếu

- Tăng nhiễu

T<sub>low</sub> quá cao

- Làm đứt gãy biên

*Tỉ lệ thường dùng:

 $$T_{low} \approx (0.3 \sim 0.5) \times T_{high}$$


### Ý nghĩa của các tham số
* **$T_{high}$ (Upper Threshold):** Xác định "cạnh mạnh". Nếu Gradient > $T_{high}$, chắc chắn là cạnh.
* **$T_{low}$ (Lower Threshold):** Xác định "cạnh yếu". Nếu Gradient < $T_{low}$, chắc chắn loại bỏ.

### Cơ chế hoạt động

Việc phân loại điểm ảnh dựa trên Gradient được thực hiện theo 3 quy tắc sau:

1. **Nếu Gradient > $T_{high}$**: Chắc chắn là cạnh (Strong Edge).
2. **Nếu Gradient < $T_{low}$**: Chắc chắn không phải là cạnh và sẽ bị loại bỏ.
3. **Nếu $T_{low} < \text{Gradient} < T_{high}$**:
    - Đây là cạnh yếu (Weak Edge).
    - Chỉ được giữ lại nếu nó **kết nối trực tiếp** với một điểm ảnh là "cạnh mạnh".
