## **THỰC HIỆN THUẬT TOÁN CANNY BẰNG CÁC THƯ VIỆN:**

```python
# sử dụng thuật toán cv2.canny để phát hiện cạnh trong hình ảnh
import cv2 
from matplotlib import pyplot as plt

# Sửa đường dẫn tương đối từ vị trí notebook
img = cv2.imread('../../../assets/images/anh_mau.jpg', cv2.IMREAD_GRAYSCALE)
plt.rcParams['figure.figsize'] = [15, 10]

# ngưỡng phổ biến hay sử dụng nhất và giúp ảnh rõ nét.
t_lower = 50;
t_upper = 150;

# Giải thích: img là ảnh đầu vào, t_lower và t_upper là hai ngưỡng để phát hiện cạnh
edges = cv2.Canny(img, t_lower, t_upper)

# Hiển thị ảnh gốc và ảnh sau khi phát hiện cạnh
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Ảnh gốc')
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Ảnh cạnh')
plt.show()
```

Ở đoạn code thấy được rằng ngưỡng `50,150` giúp ta phát hiện biên và tạo ra các edges như [iput](../input_cau1.jpg) đã tạo ra.

**Tiếp theo sẽ đưa ra:**
- Hàm Canny() với Aperture_size
- Hàm Canny() với L2Gradient
- Hàm Canny() với cả 2 Aperture size and L2gradient

```python
# Hàm Canny() với Aperture_size
import cv2 
from matplotlib import pyplot as plt

# Sửa đường dẫn tương đối từ vị trí notebook
img = cv2.imread('../../../assets/images/anh_mau1.jpg', cv2.IMREAD_GRAYSCALE)
plt.rcParams['figure.figsize'] = [15, 10]

# ngưỡng phổ biến hay sử dụng nhất và giúp ảnh rõ nét.
t_lower = 100;
t_upper = 200;

# Giải thích: img là ảnh đầu vào, t_lower và t_upper là hai ngưỡng để phát hiện cạnh
edges = cv2.Canny(img, t_lower, t_upper, apertureSize=3)

# Hiển thị ảnh gốc và ảnh sau khi phát hiện cạnh
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Ảnh gốc')
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Ảnh cạnh')
plt.show()
```

**Kết luận**: `Aperture_size` của sobel dùng để tính gradient trong `Canny`, kernel lớn sẽ giúp cho hình được mịn và nhạy hơn khi sử dụng với nó nhưng càng lớn thì biên càng lớn nên thường mất đi các chi tiết nhỏ.

```python
#Hàm Canny() với L2Gradient
import cv2 
from matplotlib import pyplot as plt

# Sửa đường dẫn tương đối từ vị trí notebook
img = cv2.imread('../../../assets/images/anh_mau2.jpg', cv2.IMREAD_GRAYSCALE)
plt.rcParams['figure.figsize'] = [15, 10]

# ngưỡng phổ biến hay sử dụng nhất và giúp ảnh rõ nét.
t_lower = 100;
t_upper = 200;

# Giải thích: img là ảnh đầu vào, t_lower và t_upper là hai ngưỡng để phát hiện cạnh
edges = cv2.Canny(img, t_lower, t_upper,L2gradient=False)

# Hiển thị ảnh gốc và ảnh sau khi phát hiện cạnh
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Ảnh gốc')
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Ảnh cạnh')
plt.show()
```

**Kết luận**: sau khi áp dụng `L2gradient` thì chúng ta sẽ thấy rõ được rằng nếu ta sử dụng `Flase` thì ảnh sẽ nhanh hơn nhưng lại không chính xác cao nhưng nếu ngược lại sử dụng `True` thì ảnh sẽ chính xác hơn nhưng lại chậm hơn khi load

**Tiếp theo là ghép cả hai lại:**

```python
#Hàm Canny() với L2Gradient và Aperture_size
import cv2 
from matplotlib import pyplot as plt

# Sửa đường dẫn tương đối từ vị trí notebook
img = cv2.imread('../../../assets/images/anh_mau.jpg', cv2.IMREAD_GRAYSCALE)
plt.rcParams['figure.figsize'] = [15, 10]

# ngưỡng phổ biến hay sử dụng nhất và giúp ảnh rõ nét.
t_lower = 100;
t_upper = 200;


# Giải thích: img là ảnh đầu vào, t_lower và t_upper là hai ngưỡng để phát hiện cạnh
edges = cv2.Canny(img, t_lower, t_upper, apertureSize=7, L2gradient=True)

# Hiển thị ảnh gốc và ảnh sau khi phát hiện cạnh
plt.subplot(121), plt.imshow(img, cmap='gray')
plt.title('Ảnh gốc')
plt.subplot(122), plt.imshow(edges, cmap='gray')
plt.title('Ảnh cạnh')
plt.show()
```
