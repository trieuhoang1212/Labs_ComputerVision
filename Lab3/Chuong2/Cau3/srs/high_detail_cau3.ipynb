```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show(title, image):
    plt.figure(figsize=(5,5))
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
img = cv2.imread("detail.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

show("Gray Image", gray)
```

```python
#lm m nh
blur = cv2.GaussianBlur(gray, (5, 5), 1)
show("Gaussian Blur", blur)
```

```python
#canny vi thresholds thp
edges_low = cv2.Canny(blur, 30, 120)
show("Canny with low thresholds", edges_low)
```

```python
#tng thresholds
low=80
high=200
edges_high = cv2.Canny(blur, low, high)
show("Canny with higher thresholds", edges_high)
```
